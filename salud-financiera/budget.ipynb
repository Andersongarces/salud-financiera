```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
expenses = [20000, 18000, 22000, 24000, 27000, 28000, 28000, 28000, 28000, 30500, 31000, 32000]  

income = [25000, 28000, 30000, 29000, 22000, 24500, 25000, 27800, 29100, 36000, 39000, 42000] 

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept','Oct', 'Nov', 'Dec']  

budget_limit = 25000  

df = pd.DataFrame({
    'Months': months,
    'Expenses': expenses, 
    'Income': income
})
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df['Months'], df['Income'], color='green', marker='o', label='Income')
ax.plot(df['Months'], df['Expenses'], color='red', marker='o', label='Expenses')

ax.hlines(y=budget_limit, xmin=0, xmax=12, colors='blue', linestyles='--', label='Budget Limit')

ax.set_xlabel('Months')
ax.set_ylabel('Amount')
ax.set_title('Monthly Income vs Expenses', fontsize=18)

ax.fill_between(df['Months'], df['Expenses'], budget_limit, 
               where=(df['Expenses'] > budget_limit), facecolor='orange', alpha=0.5, interpolate=True)
               
ax.legend(loc='best')                

plt.tight_layout()
plt.show()
```

## Análsis financiero

Para analizar la salud financiera de esta empresa, podemos calcular algunas métricas clave utilizando los datos proporcionados. Aquí hay algunas métricas que podríamos considerar:

Margen de beneficio: Esto indica cuánto de cada unidad monetaria de ingresos se convierte en beneficio neto. Se calcula restando los gastos totales de los ingresos totales y luego dividiendo el resultado por los ingresos totales.

Ratio de gastos a ingresos: Este ratio muestra la proporción de los gastos en relación con los ingresos. Se calcula sumando los gastos totales y dividiéndolos por los ingresos totales.
Desviación de gastos con respecto al presupuesto: Esto indica cuánto los gastos exceden el límite de presupuesto establecido. Se calcula tomando la diferencia entre los gastos totales y el límite de presupuesto.

```python
# Calculating metrics
df['Profit Margin'] = (df['Income'] - df['Expenses']) / df['Income']
df['Expense-Income Ratio'] = df['Expenses'] / df['Income']
df['Budget Excess'] = df['Expenses'] - budget_limit

# Displaying the results
print(df)
print("\n")

# Summary statistics
print("Summary Statistics:")
print(df.describe())
```

## Margen de beneficio

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Months'], df['Profit Margin'], marker='o', color='blue', linestyle='-')

# Línea de presupuesto
plt.axhline(y=budget_limit/df['Income'].max(), color='red', linestyle='--', label='Budget Limit')

# Añadir etiquetas y título
plt.xlabel('Months')
plt.ylabel('Profit Margin')
plt.title('Profit Margin Over Months')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Basado en el margen de beneficio representado en el gráfico, podemos hacer varias observaciones y análisis:

1. **Tendencia general del margen de beneficio:** Podemos observar si el margen de beneficio se mantiene estable, aumenta o disminuye a lo largo de los meses. Si el margen de beneficio es consistente o aumenta, puede indicar una salud financiera sólida. Por otro lado, si disminuye, podría señalar problemas financieros que requieren atención.

2. **Comparación con el límite de presupuesto:** La línea punteada que representa el límite de presupuesto nos permite ver cómo el margen de beneficio se compara con este límite. Si el margen de beneficio está consistentemente por encima del límite de presupuesto, sugiere que la empresa está generando ingresos suficientes para cubrir sus gastos y aún así obtener ganancias. Sin embargo, si el margen de beneficio cae por debajo del límite de presupuesto, podría indicar que la empresa está operando con pérdidas o que sus gastos están excediendo los ingresos, lo que podría requerir medidas correctivas.

3. **Identificación de tendencias estacionales:** Dependiendo del tipo de negocio, es posible que haya fluctuaciones estacionales en el margen de beneficio. Por ejemplo, algunas empresas pueden experimentar un margen de beneficio más bajo durante ciertas temporadas del año debido a factores estacionales o ciclos de ventas. Identificar estas tendencias estacionales puede ayudar a la empresa a planificar y gestionar sus recursos de manera más efectiva.

4. **Análisis comparativo:** Además de observar el margen de beneficio en un período de tiempo determinado, también podemos comparar el margen de beneficio de un año con otro o con el de la industria en general. Esto puede proporcionar información adicional sobre el rendimiento financiero relativo de la empresa y su posición en el mercado.

En resumen, el análisis del margen de beneficio nos permite evaluar la eficiencia financiera de la empresa, identificar áreas de mejora y tomar decisiones estratégicas para mantener o mejorar su salud financiera.

1. **Tendencia general del ratio de gastos a ingresos:** Observando el gráfico, podemos ver cómo varía el ratio de gastos a ingresos a lo largo de los meses. Si el ratio de gastos a ingresos está por encima de esta línea, indica que los gastos superan el límite establecido por la empresa. Por otro lado, si está por debajo de la línea, significa que los gastos están por debajo del límite. Esta comparación permite evaluar si la empresa está operando dentro de sus límites de presupuesto establecidos.

3. **Identificación de meses con ratios atípicamente altos o bajos:** Podemos identificar meses específicos donde el ratio de gastos a ingresos es particularmente alto o bajo. Esto puede ser indicativo de períodos de eficiencia financiera excepcional o dificultades financieras que requieren atención.

4. **Análisis comparativo:** Al igual que en los análisis anteriores, sería útil comparar estos datos con los de años anteriores y con los de otras empresas del mismo sector para evaluar la eficiencia financiera relativa de la empresa y identificar áreas de mejora.

En resumen, este análisis nos permite evaluar cómo los gastos de la empresa en relación con sus ingresos fluctúan a lo largo del tiempo y compararlos con los límites de presupuesto establecidos para tomar decisiones informadas sobre la gestión financiera.

## Desviación de gastos con respecto al presupuesto:

```python

# Calcular el ratio de gastos a ingresos
df['Expense-Income Ratio'] = df['Expenses'] / df['Income']

# Calcular la desviación de gastos con respecto al presupuesto
budget_deviation = df['Expenses'].sum() - (budget_limit * len(df))

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(df['Months'], df['Expense-Income Ratio'], marker='o', color='blue', linestyle='-', label='Expense-Income Ratio')

# Línea de presupuesto
plt.axhline(y=budget_limit / df['Income'].max(), color='red', linestyle='--', label='Budget Limit')

# Añadir etiquetas y título
plt.xlabel('Months')
plt.ylabel('Expense-Income Ratio')
plt.title('Expense-Income Ratio Over Months')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Análisis de la desviación de gastos con respecto al presupuesto
print("Desviación de gastos con respecto al presupuesto:", budget_deviation)
```

El análisis de la desviación de gastos con respecto al presupuesto nos indica cuánto los gastos totales exceden o están por debajo del límite de presupuesto establecido. Si el resultado es negativo, significa que los gastos están por debajo del presupuesto, mientras que un resultado positivo indica que los gastos exceden el presupuesto. Este análisis proporciona información adicional sobre la gestión financiera de la empresa y su capacidad para controlar los gastos dentro de los límites establecidos.
