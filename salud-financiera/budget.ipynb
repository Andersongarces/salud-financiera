```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
expenses = [20000, 18000, 22000, 24000, 27000, 28000, 28000, 28000, 28000, 30500, 31000, 32000]  

income = [25000, 28000, 30000, 29000, 22000, 24500, 25000, 27800, 29100, 36000, 39000, 42000] 

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept','Oct', 'Nov', 'Dec']  

budget_limit = 25000  

df = pd.DataFrame({
    'Months': months,
    'Expenses': expenses, 
    'Income': income
})
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df['Months'], df['Income'], color='green', marker='o', label='Income')
ax.plot(df['Months'], df['Expenses'], color='red', marker='o', label='Expenses')

ax.hlines(y=budget_limit, xmin=0, xmax=12, colors='blue', linestyles='--', label='Budget Limit')

ax.set_xlabel('Months')
ax.set_ylabel('Amount')
ax.set_title('Monthly Income vs Expenses', fontsize=18)

ax.fill_between(df['Months'], df['Expenses'], budget_limit, 
               where=(df['Expenses'] > budget_limit), facecolor='orange', alpha=0.5, interpolate=True)
               
ax.legend(loc='best')                

plt.tight_layout()
plt.show()
```

## Anlsis financiero

Para analizar la salud financiera de esta empresa, podemos calcular algunas mtricas clave utilizando los datos proporcionados. Aqu hay algunas mtricas que podramos considerar:

Margen de beneficio: Esto indica cunto de cada unidad monetaria de ingresos se convierte en beneficio neto. Se calcula restando los gastos totales de los ingresos totales y luego dividiendo el resultado por los ingresos totales.

Ratio de gastos a ingresos: Este ratio muestra la proporcin de los gastos en relacin con los ingresos. Se calcula sumando los gastos totales y dividindolos por los ingresos totales.
Desviacin de gastos con respecto al presupuesto: Esto indica cunto los gastos exceden el lmite de presupuesto establecido. Se calcula tomando la diferencia entre los gastos totales y el lmite de presupuesto.

```python
# Calculating metrics
df['Profit Margin'] = (df['Income'] - df['Expenses']) / df['Income']
df['Expense-Income Ratio'] = df['Expenses'] / df['Income']
df['Budget Excess'] = df['Expenses'] - budget_limit

# Displaying the results
print(df)
print("\n")

# Summary statistics
print("Summary Statistics:")
print(df.describe())
```

## Margen de beneficio

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Months'], df['Profit Margin'], marker='o', color='blue', linestyle='-')

# Lnea de presupuesto
plt.axhline(y=budget_limit/df['Income'].max(), color='red', linestyle='--', label='Budget Limit')

# Aadir etiquetas y ttulo
plt.xlabel('Months')
plt.ylabel('Profit Margin')
plt.title('Profit Margin Over Months')
plt.legend()

# Mostrar el grfico
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Basado en el margen de beneficio representado en el grfico, podemos hacer varias observaciones y anlisis:

1. **Tendencia general del margen de beneficio:** Podemos observar si el margen de beneficio se mantiene estable, aumenta o disminuye a lo largo de los meses. Si el margen de beneficio es consistente o aumenta, puede indicar una salud financiera slida. Por otro lado, si disminuye, podra sealar problemas financieros que requieren atencin.

2. **Comparacin con el lmite de presupuesto:** La lnea punteada que representa el lmite de presupuesto nos permite ver cmo el margen de beneficio se compara con este lmite. Si el margen de beneficio est consistentemente por encima del lmite de presupuesto, sugiere que la empresa est generando ingresos suficientes para cubrir sus gastos y an as obtener ganancias. Sin embargo, si el margen de beneficio cae por debajo del lmite de presupuesto, podra indicar que la empresa est operando con prdidas o que sus gastos estn excediendo los ingresos, lo que podra requerir medidas correctivas.

3. **Identificacin de tendencias estacionales:** Dependiendo del tipo de negocio, es posible que haya fluctuaciones estacionales en el margen de beneficio. Por ejemplo, algunas empresas pueden experimentar un margen de beneficio ms bajo durante ciertas temporadas del ao debido a factores estacionales o ciclos de ventas. Identificar estas tendencias estacionales puede ayudar a la empresa a planificar y gestionar sus recursos de manera ms efectiva.

4. **Anlisis comparativo:** Adems de observar el margen de beneficio en un perodo de tiempo determinado, tambin podemos comparar el margen de beneficio de un ao con otro o con el de la industria en general. Esto puede proporcionar informacin adicional sobre el rendimiento financiero relativo de la empresa y su posicin en el mercado.

En resumen, el anlisis del margen de beneficio nos permite evaluar la eficiencia financiera de la empresa, identificar reas de mejora y tomar decisiones estratgicas para mantener o mejorar su salud financiera.

## Desviacin de gastos con respecto al presupuesto:

```python

# Calcular el ratio de gastos a ingresos
df['Expense-Income Ratio'] = df['Expenses'] / df['Income']

# Calcular la desviacin de gastos con respecto al presupuesto
budget_deviation = df['Expenses'].sum() - (budget_limit * len(df))

# Crear el grfico
plt.figure(figsize=(10, 6))
plt.plot(df['Months'], df['Expense-Income Ratio'], marker='o', color='blue', linestyle='-', label='Expense-Income Ratio')

# Lnea de presupuesto
plt.axhline(y=budget_limit / df['Income'].max(), color='red', linestyle='--', label='Budget Limit')

# Aadir etiquetas y ttulo
plt.xlabel('Months')
plt.ylabel('Expense-Income Ratio')
plt.title('Expense-Income Ratio Over Months')
plt.legend()

# Mostrar el grfico
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Anlisis de la desviacin de gastos con respecto al presupuesto
print("Desviacin de gastos con respecto al presupuesto:", budget_deviation)
```

El anlisis de la desviacin de gastos con respecto al presupuesto nos indica cunto los gastos totales exceden o estn por debajo del lmite de presupuesto establecido. Si el resultado es negativo, significa que los gastos estn por debajo del presupuesto, mientras que un resultado positivo indica que los gastos exceden el presupuesto. Este anlisis proporciona informacin adicional sobre la gestin financiera de la empresa y su capacidad para controlar los gastos dentro de los lmites establecidos.
